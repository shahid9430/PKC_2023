# Outliers removal
Visualization method

```python
# import libraries
import seaborn as sns
```

```python
sns.boxplot(data=df, y='age', x='sex')
```

```python
# IQR Method
Q1= df.age.quantile(0.25)
Q3= df.age.quantile(0.75)
IQR= Q3-Q1
lower_bound= Q1- 1.5*IQR
upper_bound= Q3 + 1.5*IQR
df= df[(df['age']> lower_bound)& (df['age']< upper_bound)]
```

```python
df.shape
```

```python
Q1= df.fare.quantile(0.25)
Q3= df1.fare.quantile(0.75)
IQR= Q3-Q1
lower_bound= Q1- 1.5*IQR
upper_bound= Q3 + 1.5*IQR
df= df[(df['fare']> lower_bound)& (df['fare']< upper_bound)]
```

```python
df.shape
```

```python
df['age'].std()
```

```python
# z-score method:
import pandas as np
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
from scipy import stats
```

```python
zscore = np.abs(stats.zscore(df['age'])) 
threshold = 3
df= df[(zscore < threshold).all(axis='age')]
```

# **Log Transformation**

```python
# load data
import pandas as pd
kashti=sns.load_dataset('titanic')
```

```python
kashti.head()
```

```python
kashti.info()
```

```python
import numpy as np
# fill the null values in the age cols(age, fare) of dataset
kashti['age']= kashti['age'].fillna(kashti['age'].mean())
kashti['fare']= kashti['fare'].fillna(kashti['fare'].mean())
# log transform 
kashti['age']= np.log(kashti['age'])
kashti['fare']= np.log(kashti['fare'])
```

```python
kashti.head()
```

```python
sns.boxplot(data=kashti, x='sex', y='age')
```
