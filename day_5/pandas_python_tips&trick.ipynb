# 15- Splitting a string into multiple columns

```python
# Importing libraries
import pandas as pd
df= pd.DataFrame({'name': ['akhar ali','gul nawaz', 'wahid khan', 'shawal khwash'],
                  'location': ['walgaii, Ambrika', 'kotkai, Iran','topiklai, India', 'sagai, Afgha']})
df
```

```python
# splitting a column
df.name.str.split(' ', expand=True)
```

```python
# adding the splitting into new col
df[['first_name', 'last_name']]= df.name.str.split(' ', expand=True)
df
```

```python
df.location.str.split(' ', expand=True)
```

```python
df[['province', 'state']]= df.location.str.split(' ', expand=True) # if  we write (', '), then the 
# common in the below table will not show
df
```

```python
# refine data manipulation
df= df[['first_name', 'last_name', 'province', 'state']]
df
```

# 21- Reshaping a Data frame

```python
# we are creating a dataframe a nd giving it a name distance
distance= pd.DataFrame([['12345',100,200,300],['23456',400,500,600],['45678', 500,600,700],[56789, 600,700,800]], 
                        columns=['zip','factory', 'warehouse', 'retail'])
distance.head()
```

```python
# if we want to draw its bar plot
sns.barplot(x='zip', y='factory', data=distance)
```

```python
distance.head().T
# we do convert the dataframe to long formate for a better plot formation and understanding, in a short formate we cannot put hue value
```

```python
#but we want the data to be of the folloeing type
distance_2 = pd.DataFrame([[1,'12345', 'factory'], [2, '23456', 'warehouse']], 
                          columns=['user_id', 'zip', 'location_type'])
distance_2.head()
```

```python
# so to arrange the 'distance' dataframe data of the above type we will do
distance
```

```python
distance_1= distance.melt(id_vars='zip', var_name='location_type', value_name='fasla')
distance_1 
```
