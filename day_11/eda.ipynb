# EDA
## Exploratory Data Analysis: It is the method to analyse and summarise the datasets.

```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# info function gives the dtype and shape function ata time
df.info()
```

```python
#summarize (summary statistics)
df.describe()
```

```python
# find null values
df.isnull().sum()
```

```python
# to check the % of null values
df.isnull().sum()*100//len(df)
```

```python
# finding unique values( nunique mean number of unique values)
df.nunique()
```

```python
# visualize the NaN values in heatmap
sns.heatmap(df.isnull())
```

```python
# to make the null values more visible and understanable
sns.heatmap(df.isnull(), cmap='viridis', yticklabels= False, cbar=False)
```

> the yellow dashes are the null/ NaN values in the corresponding cols

```python
df.drop(columns=('deck'))
```

```python
#visualize and explore
sns.countplot(data=df, x="survived")
#in the countplot we can not give y-axis because its alreday occupied by count
```

```python
# y is given in the bar plot
sns.barplot(data=df, x="survived", y="age", hue="who")
```

```python
# check data distribution
sns.histplot(data=df, x='age')
```

```python
# bins are the bars of plot, to reduce or increase it we give it numbers manually
sns.histplot(data=df, x='age', bins=5)
```

```python
sns.histplot(data=df, x='age', bins= 20)
```

```python
sns.histplot(data=df, x='age', color='darkred', alpha=0.7)
# alpha is used for color saturation/opacity
```

```python
# the above graph has no grid, let draw through another way
df['age'].hist(bins=30, color= 'darkred', alpha= 0.7)
```

```python
sns.boxplot(data=df, y="age") # gave the age on y-axis so the box will allign itself to y-axis
```

```python
# if we plot fare with the age
sns.boxplot(data=df, x="age", y="fare")
```

In the above plot the boxes are saturated because the unique

 values of
 fare are so much, if we draw age with survived *survived have two unique values*

```python
sns.boxplot(data=df, y="age", x="survived")
#  this resulted in two box plots because survived has teo unique values
```

In the boxplot we remove the outliers


***While plotting ***

    Plotting barplot/box plot: we take categorical variable vs numerical var(on y-axis)
    plotting countplot we put only one categorical variable
    plotting histogram we put only one numeric variable
    If we want to plot two numeric variable we draw scatter plot


```python
sns.boxenplot(data=df, y="age", x="survived", hue='who')
```

```python
# to draw to numeric variable we plot scatter plot
sns.scatterplot(data=df, x="age", y="fare")
```

```python
#in the above plot y ranges upto 500, we want to limit it to 300
sns.scatterplot(data=df, x="age", y="fare")
plt.ylim(0,300)
```

If we apply hue= pclass, ( because fare changes with respect to the pclass)

```python
sns.scatterplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0,300)
```

```python
#lets check the pairplot, hue = survived, 
sns.pairplot(data=df, hue="survived")
```

```python
#stacked barchart/plot
df.groupby(["pclass",'who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarm plot
sns.swarmplot(data=df, x="age", y="fare")
```

```python
# let's change some variable

sns.swarmplot(data=df, x="pclass", y="age")
```

```python

sns.swarmplot(data=df, x="pclass", y="age", hue="survived")
```

```python

sns.swarmplot(data=df, x="survived", y="age", hue="who")
```

```python
#point plot/ catplot

sns.catplot(data=df, x="pclass", y="fare", hue="embarked", kind='point')
```

```python
sns.lmplot(data=df[df['fare']<300], x="age", y="fare")
```

```python
# relation between two numeric variables/ LM plot
sns.lmplot( x="age", y= "fare", hue="pclass", data=df)
```

```python
# how to add a column and make an age group
sns.histplot(data=df, x="age")
```

```python
# grouping of the age
df.columns
```

```python
#  to add a new column
df['age_group']= pd.cut(df['age'], bins=[0,10,18,30,50,80],
 labels=['Kids', 'Child', 'Young age', 'Middle Age', 'Old Age'])
```

```python
df.head()
```

```python
df.groupby("age_group")['survived'].mean().plot(kind= 'bar')
```

```python
# to make a pairplot of the variables(vars ) of your choice
sns.pairplot(data=df, vars=['age', 'fare', 'pclass'], hue='survived')
```
