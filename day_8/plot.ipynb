```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kashti=sns.load_dataset('titanic')
```

```python
kashti
```

```python
kashti.info()
```

```python
# before plotting we check null values
kashti.isnull().sum()
```

```python
# to make bar plot of the survived
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x="survived", data= kashti)
```

```python
sns.countplot(x='survived', hue='sex', data=kashti)
```

```python
# to draw a scatter plt
plt.scatter(data=kashti, x="age", y="fare")
```

```python
# to apply colors to the plot we can't use hue function in matplotlib, instead we use color function
plt.scatter(data=kashti, x="age", y="fare", color="red")
```

```python
# to draw the same scatter plot through seaborn
sns.scatterplot(data=kashti, x="age", y="fare", hue='pclass')
```

```python
sns.scatterplot(data=kashti, x="sex", y="fare", hue='pclass')
```

```python
# the upper graph is not that much sophisticated we instead draw a bar graph
sns.barplot(data=kashti, x="sex", y="fare", hue='pclass')
```

```python
#pairplot draw a graph between each col of the data
sns.pairplot(kashti)
```

```python
# to draw a histogram
kashti.hist()
```

```python
# bins are the bars of the graph, we can draw them in required numbers
kashti.hist(bins=10)
```

```python
# also we can reashpe the histogram
kashti.hist(bins=10, figsize=(9,7))
```

```python
sns.boxplot(data=kashti, x="sex", y="fare", hue='parch')
```

```python
sns.boxplot(data=kashti, x
    = "sex", y="fare", hue='embark_town')
```

```python
sns.boxenplot(data=kashti, x="sex", y="fare", hue='pclass')
```

```python
# grid plot in sns. the programme is copied
g= sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True, 
              palette={1: "blue", 0:"red"})
g= g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

```python
# strip plot is actually a combination of bar and scatter plot
sns.stripplot(data=kashti, x="sex", y="fare", hue='class')
```
