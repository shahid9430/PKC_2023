```python
# to remove a col, in this case the unnamed: 0 col
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

```python
# to separate the male female data we use groupby function
df.groupby('sex').describe()
```

```python
# to transpode it
df.groupby('sex').describe().T
```

```python
#count the number of survived =1 grouped by sex (male or female)
df.groupby('sex').survived.count()
```

```python
#another way to view as 
df.groupby(['survived','sex']).survived.count()
```

```python
df.groupby(['sex','survived','who']).who.count()
```

```python
# to plot it
sns.countplot(x='survived', data= df)
```

```python
sns.countplot(x='survived', hue='sex', data=df)
```
